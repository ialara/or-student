```python
#Sim #4: Infant mortatlity added
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
#Constants
init_start_age = 18
init_end_age = 50

fertility_start = 18
fertility_end = 40
baby_odds = 0.25

old_person = 60
old_death_odds = 0.1
```

```python
class person: #Added
    def __init__(self, age):
        self.gender = random.randint(0,1)
        self.age = age

def begin_sim():
    for individual in range(int(population[0])):
        peopleDictionary.append(person(random.randint(init_start_age, init_end_age)))

def run_year():
    deaths = 0
    births = birth()
    for individual in peopleDictionary:
        if individual.age > old_person:
            if random.uniform(0,1) <= old_death_odds:
                peopleDictionary.remove(individual)
                deaths += 1
        else:
            individual.age +=1
    return births, deaths
```

```python
#Constants Added
infant_mortality_reduction = 0.985 #Added

#Variable initialization
year = 0
infant_mortality = 0.62 #Added

population = np.zeros(10000)
population[0] = 50
peopleDictionary = []

def birth(): #Added
    births = 0
    for individual in peopleDictionary:
        if individual.gender == 1 and \
           individual.age > fertility_start and \
           individual.age < fertility_end and \
           random.uniform(0,1) <= baby_odds * (1 - infant_mortality): #Added
              peopleDictionary.append(person(0))
              births += 1
              #print(len(peopleDictionary))
    return births

random.seed(180159)
begin_sim()
while population[year] < 100000 and population[year] > 1:
    run_year()
    infant_mortality *= infant_mortality_reduction
    year += 1
    population[year] = len(peopleDictionary)

years=np.arange(10000)
plt.xlim([0,200])
plt.ylim([0,200])
plt.plot(years[0:year],population[0:year])
print('Population exceeded 1,000,000 in year', year)
```

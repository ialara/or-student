```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
#Constants
darts = 100000

#Arrays
red_dots = np.zeros((2, darts))
blue_dots = np.zeros((2, darts))

num_red = 0
num_blue = 0
for dart in range(darts):
    x = random.uniform(-1, 1)
    y = random.uniform(-1, 1)
    if x*x + y*y <= 1:
        red_dots[0][num_red] = x
        red_dots[1][num_red] = y
        num_red += 1
    else:
        blue_dots[0][num_blue] = x
        blue_dots[1][num_blue] = y
        num_blue += 1

print(4 * num_red/darts)
plt.scatter(red_dots[0], red_dots[1], c = 'red', s = 0.01, marker = 'p')
plt.scatter(blue_dots[0], blue_dots[1], c = 'blue', s = 0.01, marker = 'p')
```

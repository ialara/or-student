```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def get_iter(c:complex, thresh, max_steps) -> int:
    z = c
    i = 1
    while i < max_steps and (z*z.conjugate()).real < thresh:
        z = z*z + c
        i += 1
    return i

def plotter(x_start, x_len, x_steps, y_start, y_len, y_steps, thresh, max_steps):
    mx = x_len / (x_steps-1)
    my = y_len / (y_steps-1)
    mapper = lambda x,y: (mx*x + x_start, my*y + y_start)
    img=np.full((y_steps, x_steps), 255)
    for x in range(x_steps):
        for y in range(y_steps):
            it = get_iter(complex(*mapper(x,y)), thresh, max_steps)
            img[y][x] = 255 - it
    return img

x_start = -2.1
x_len = 2.9
x_steps = 2000
y_start = -1.2
y_len = 2.4
y_steps = 1000
threshold = 4
max_steps = 50
img = plotter(x_start, x_len, x_steps, y_start, y_len, y_steps, threshold, max_steps)
plt.imshow(img, cmap="plasma", aspect="auto")
plt.axis("off")
plt.show()
```

```python
import numpy as np
from numpy.random import default_rng
import math
import matplotlib.pyplot as plt

rng = default_rng()

def make_vertices(sides):
    vx = np.empty(0)
    vy = np.empty(0)
    angle = math.pi/2
    angle_delta = 2*math.pi/sides
    for s in range(sides):
        vx = np.append(vx, np.cos(angle+s*angle_delta))
        vy = np.append(vy, np.sin(angle+s*angle_delta))
    vx = np.append(vx,0)
    vy = np.append(vy,1)
    return vx, vy

def play_game(sides, points):
    vx, vy = make_vertices(sides)
    x = np.empty(0)
    y = np.empty(0)
    r = np.empty(0)

    x = np.append(x, vx[0])
    y = np.append(y, vy[0])
    r = np.append(r, 0)

    for point in range(points):
        v = rng.integers(0, sides)
        #if v >= r[point]: v = (v + 1) % sides
        x = np.append(x, (x[point] + vx[v])/2)
        y = np.append(y, (y[point] + vy[v])/2)
        r = np.append(r, v)
    return vx, vy, x, y

sides = 3
points = 10000
vx, vy, x, y = play_game(sides, points, )

plt.figure(figsize=(8,8))
plt.plot(vx, vy)
plt.scatter(x,y, s=1)
plt.show()
```
